```python
import pandas as pd

print("Starting conversion...")

# 1. Process u.data (Ratings)
# This file is TAB-separated
r_cols = ['user_id', 'item_id', 'rating', 'timestamp']
df_ratings = pd.read_csv(
    'u.data', 
    sep='\t', 
    names=r_cols, 
    header=None
)
# Save as CSV
df_ratings.to_csv('ratings.csv', index=False)
print("Saved ratings.csv")
```

```python
import pandas as pd

print("Starting MovieLens 1M conversion...")

# Define the column names
r_cols = ['UserID', 'MovieID', 'Rating', 'Timestamp']

# Read the file
# We must use engine='python' because the separator '::' is not a single character
df_ratings_1m = pd.read_csv(
    'ratings.dat',
    sep='::',
    names=r_cols,
    header=None,
    engine='python'  # This is the important part!
)

# Save as CSV
df_ratings_1m.to_csv('ratings_1m.csv', index=False)

print("Saved ratings_1m.csv successfully!")
```
